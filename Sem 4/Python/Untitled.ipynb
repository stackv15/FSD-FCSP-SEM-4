```python
import pandas as pd
import numpy as np
data={
    "Name":["Ravi",'Kishor','Jay',np.NaN,np.NaN,'Viki'],
    "Salary":[20000,30900,np.NaN,np.NaN,62000,np.NaN],
    "expect":[np.nan,10900,30000,np.NaN,np.NaN,40000],
    "region":["Bopal","Shyamal",np.NaN,np.NaN,"Maninagar","Vasna"]
}
dataset=pd.DataFrame(data)
dataset
```

```python
dataset.isna()
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(axis=0,how='all',thresh=3)
```

```python
dataset.dropna(subset=['Salary','expect','region'])
```

```python
dataset
```

```python
dataset.dropna(axis=0,how='all',thresh=3,inplace=True)
```

```python
dataset
```

```python
import pandas as pd
import numpy as np
data={
    "Name":["Ravi",'Kishor','Jay',np.NaN,np.NaN,'Viki'],
    "Salary":[20000,30900,np.NaN,np.NaN,62000,np.NaN],
    "expect":[np.nan,10900,30000,np.NaN,np.NaN,40000],
    "region":["Bopal","Shyamal",np.NaN,np.NaN,"Maninagar","Vasna"]
}
dataset=pd.DataFrame(data)
dataset
```

```python
dataset.fillna(0)
```

```python
dataset.describe()
```

```python
#FILL THE NULL VALUE BY USING MEAN AND MEDIAN VALUE
```

```python
dataset['Salary'].fillna(dataset['Salary'].mean())
```

```python
dataset['Salary'].fillna(dataset['Salary'].median())
```
