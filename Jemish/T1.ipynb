```python
import pandas as pd
```

```python
import numpy as np
```

```python
#List
data = [1,32,33,42]
pd.Series(data)
```

```python
#Tuple
data = (23,4,54,6,66)
pd.Series(data)
```

```python
#Dictionary
data = {"Jemish" : 24, "Yug" : 25}
pd.Series(data)
```

```python
#Asign the Index
data = [1,32,33,42]
pd.Series(data,index=["A","B","C","D"])
```

```python
# Series using numpy array
data = np.array([1,2,3,4])
pd.Series(data)
```

```python
#change the data type of Series
data = np.array([1,2,3,4])
pd.Series(data,dtype="float")
```

```python
#Scaler value
pd.Series(5,index=[1,2,3,4])
```

```python
#
pd.Series(5,index=[1,2,3,4],dtype="string",name="Number")
```

```python
data = [10,11,12,13,14,15]
ser = pd.Series(data,index=["A","B","C","D","E","F"],dtype="float",name="Num")
print(ser.values)
```

```python
print(ser.index)
print(ser.dtype)
print(ser.size)
print(ser.shape)
print(ser.name)
print(ser.is_unique)
print(ser.min())
print(ser.max())
```

```python
data = [10,20,35,40,45,60]
s = pd.Series(data,index=["a","b","c","d","e","f"])
print(s.loc["c"])
print(s.loc["b":"e"])
print(s.iloc[3])
```

```python
data = [10,20,35,40,45,60]
s = pd.Series(data)
print(s)
print(s.loc[1])
```

```python
data = [10,20,35,40,45,60]
s = pd.Series(data,index=[1,2,3,4,5,6])
print(s.loc[2:6])
```

```python
print(s.iloc[0:4])
```

```python
data = [10,20,35,40,45,60]
s = pd.Series(data,index=["a","b","c","d","e","f"])
v1 = s.loc["d"] = 100
print(s)
```

```python
data = [10,20,35,40,20,60]
inx = ["a","b","c","d","e","f"]
s = pd.Series(data,index=inx)
for i in inx:
    if s.loc[i] > 20:
        print(f"{i} : {s.loc[i]}")
```

```python
#Data Fram
df = pd.DataFrame([[1,2,3,4],
                ["a","b","c","d"],
                [70,65,45,85]],
                index = ["ID","NAME","MARKS"],
                columns = ['x','y','z','w'] )
print(df)
```
